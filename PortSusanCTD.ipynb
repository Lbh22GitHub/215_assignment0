**3. Run the python code to read variables from the CTD (.cnv) files**

```python
# This cell loads the needed packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading Salinity at Depth
theFile = data_filepath+'106600701.cnv'
z7, S7 = np.genfromtxt(fname=theFile,usecols=(1,4),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600801.cnv'
z8, S8 = np.genfromtxt(fname=theFile,usecols=(1,4),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600901.cnv'
z9, S9 = np.genfromtxt(fname=theFile,usecols=(1,4),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106601001.cnv'
z10, S10 = np.genfromtxt(fname=theFile,usecols=(1,4),skip_header=262,unpack=True, encoding='iso-8859-1')
```

```python
#Plotting Salinity at Depth
# Some parameters for the plot as a whole
plt.rcParams["figure.figsize"] = (8, 10)
plt.rcParams["legend.loc"] = 'upper left'
fig, ax = plt.subplots()

# Modify these lines to plot different variables or add lines to add more curves
ax.plot(S7,z7)
ax.plot(S8,z8)
ax.plot(S9,z9)
ax.plot(S10,z10)
ax.set(ylabel='Depth (m)',xlabel='Salinity (psu)', title='A')

# Makes depth increase downwards
ax.invert_yaxis()

# Don't forget to change the legend and plot file name as needed!
ax.legend(('Sta7','Sta8','Sta9','Sta10'))
plt.savefig(dir_filepath+'SalinityProfiles.png')
plt.show()
```

```python
# This cell shows an example of how to import depth (z) and temperature (T) from columns 1 and 3 (for just two stations)
# You can open the .cnv file in a text editor to see the order of columns
theFile = data_filepath+'106600701.cnv'
z7, T7 = np.genfromtxt(fname=theFile,usecols=(1,3),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600801.cnv'
z8, T8 = np.genfromtxt(fname=theFile,usecols=(1,3),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600901.cnv'
z9, T9 = np.genfromtxt(fname=theFile,usecols=(1,3),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106601001.cnv'
z10, T10 = np.genfromtxt(fname=theFile,usecols=(1,3),skip_header=262,unpack=True, encoding='iso-8859-1')
```

**4. Run code to plot some stuff and save the plot file to your outer directory**

```python
#Loading Temperature at Depth
# Some parameters for the plot as a whole
plt.rcParams["figure.figsize"] = (8, 10)
plt.rcParams["legend.loc"] = 'upper left'
fig, ax = plt.subplots()

# Modify these lines to plot different variables or add lines to add more curves
ax.plot(T7,z7)
ax.plot(T8,z8)
ax.plot(T9,z9)
ax.plot(T10,z10)
ax.set(ylabel='Depth (m)',xlabel='Temperature (deg C)', title='B')

# Makes depth increase downwards
ax.invert_yaxis()

# Don't forget to change the legend and plot file name as needed!
ax.legend(('Sta7','Sta8','Sta9','Sta10'))
plt.savefig(dir_filepath+'TemperatureProfiles.png')
plt.show()


```

```python
#Loading DO at Depth
theFile = data_filepath+'106600701.cnv'
z7, O7 = np.genfromtxt(fname=theFile,usecols=(1,6),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600801.cnv'
z8, O8 = np.genfromtxt(fname=theFile,usecols=(1,6),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106600901.cnv'
z9, O9 = np.genfromtxt(fname=theFile,usecols=(1,6),skip_header=262,unpack=True, encoding='iso-8859-1')
theFile = data_filepath+'106601001.cnv'
z10, O10 = np.genfromtxt(fname=theFile,usecols=(1,6),skip_header=262,unpack=True, encoding='iso-8859-1')
```

```python
#Plotting DO at Depth
# Some parameters for the plot as a whole
plt.rcParams["figure.figsize"] = (8, 10)
plt.rcParams["legend.loc"] = 'upper left'
fig, ax = plt.subplots()

# Modify these lines to plot different variables or add lines to add more curves
ax.plot(O7,z7)
ax.plot(O8,z8)
ax.plot(O9,z9)
ax.plot(O10,z10)
ax.set(ylabel='Depth (m)',xlabel='Dissolved Oxygen (umol/kg)',title = 'C')
# Makes depth increase downwards
ax.invert_yaxis()

# Don't forget to change the legend and plot file name as needed!
plt.savefig(dir_filepath+'DOProfiles.png')
plt.show()
```
